### Adding Program School Name

```python
program_school_name = ["SoME", "SOES", "SOM", "SoAS", "SOL"]

schools_lst = []
for i in range(498):
    schools_lst.append("entry.767675694=" + np.random.choice(program_school_name))

schools_lst
```

### Are you a Gamer?

```python
areYouAGamer = ["Yes", "No"]

areYouAGamer_lst = []
for i in range(498):
    areYouAGamer_lst.append("entry.1534863658=" + np.random.choice(areYouAGamer, p = [0.4, 0.6]))

areYouAGamer_lst
```

### What Type of Gamer? 

```python
whatTypeOfGamer = ["PC", "Mobile", "Console"]

whatTypeOfGamer_lst = []
for i in range(498):
    whatTypeOfGamer_lst.append("entry.1274209221=" + np.random.choice(whatTypeOfGamer, p = [0.3, 0.4, 0.3]))

whatTypeOfGamer_lst
```

### Which Game to play in Experience Zone ?

```python
gameToPlayInExperienceZone = ["VR", "Play station", "Lan Gaming"]

gameToPlayInExperienceZone_lst = []
for i in range(498):
    gameToPlayInExperienceZone_lst.append("entry.759405699=" + np.random.choice(gameToPlayInExperienceZone, p = [0.5, 0.25, 0.25]))

gameToPlayInExperienceZone_lst
```

### Correcting Email Format as per Automating Df

```python
df["email"] = df["email"].apply(lambda x: "emailAddress=" + x)
```

### Correcting Contact Format as per Automating Df

```python
# Converting contact column to int datatype

df["contact"] = df["contact"].astype("str")

df["contact"] = df["contact"].apply(lambda x: "entry.1585933220=" + x)
```

### Adding Other Data to the Column

```python
df["Are you a Gamer?"] = areYouAGamer_lst
df["What Type of Gamer?"] = whatTypeOfGamer_lst
df["Choose game to play in Experience Zone"] = gameToPlayInExperienceZone_lst
df["School Name"] = schools_lst
```
